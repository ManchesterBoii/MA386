```python
#Lesson 9, Problem 1

def readcsv(filename,numHeader=1,dtFormat="%m/%d/%Y"):
    """  
    readcsv() is a function that reads data from a csv file and assigns
    it to a nested list data structure.
    Inputs: fileName - is a string without the .csv (which is assumed);
            numHeader - is an integer that gives the number of lines of headers
                     before the data starts;
            dtFormat - is either not assigned if there are no dates
                     in the data or a date format such as '%m/%d/%Y' as an
                     example of a date format.  The date must also be in the
                     first column for this function to work properly.
    Outputs: head - header lines as one large text
             data - the data from the columns that comes after the header
    Example: 
           h, data = readcsv("name",dtFormat = "%Y-%m-%d")
    """
    import datetime as dt
    infile = open(filename+".csv", "r")
    header = []
    for j in range(numHeader):
        h = infile.readline().split(',')
        h[-1] = h[-1].strip()
    data = []
    for line in infile:
        row = line.split(",")
        row[-1] = row[-1].strip()
        for i,e in enumerate(row):
            try:
                row[i] = dt.datetime.strptime(e,dtFormat)
            except:
                try:
                    row[i] = int(row[i])
                except:
                    try:
                        row[i] = float(row[i])
                    except:
                        pass
        data.append(row)
    infile.close()
    return(header,data)
```

```python
def writecsv(header, data, csvName, decimals):
    """
    writecsv() writes a set of data and header to a csv file with specific
    column widths - most of the time the col_width = 0

    Inputs: header - list containing column header text that goes at the top of the file
            data - nested list of the data that will be written to the file
            csv_filename - the name of the text file where the data is written
            decimals - the number of decimals for the data being written

    Outputs: nothing is returned to the program when this function is called
            csv_filename.csv is created with the header and data in it.
    """
    outfile = open(csvName + '.csv', 'w')
    for i in range(len(header)):
        outfile.write(header[i])
        if i != len(header) - 1:
            outfile.write(',')
        else:
            outfile.write('\n')
    fmt = '.' + str(decimals) + 'f'
    for i in range(len(data)):
        for j in range(len(data[i])):
            if isinstance(data[i][j], float):
                outfile.write(f"{data[i][j]:{fmt}}")
            else:
                outfile.write(data[i][j])
            if j != len(data[i]) - 1:
                outfile.write(',')
            else:
                outfile.write('\n')
    outfile.close()

import random

data = [[random.random()*random.random() for i in range(4)] for i in range(4)]
header = ["one", "two", "three", "four"]
                              
writecsv(header, data, "test", 3)
```

```python
#Lesson 9, Problem 3

def MA(data, N):
    '''
    returns an MA(N) moving average for a single vector of data
    Inputs: data - a single vector of time series data
            N - the number of time periods used to calculate the moving avg
    Outputs: a single vector that represents the MA(N) moving avg for the data. It will have N 
             fewer data points than the original vector of time series data.
    '''
    m = len(data)
    ma = []
    if N > m:
        print("Moving average not possible. Desired moving average length longer than vector.")
    for i in range(m-N+1):
        ma.append(sum(data[i:i+N])/N)
    return ma

import random
data = [random.random() for i in range(10)]

print(MA(data,5))
```

```python
def holts(data, a, b, n):
    """
    Inputs: data, alpha, beta, and N ≡ the number of time periods used for initialization.
    Output: a single vector of data that represents the forecast. 
    Function: The function accepts a data set that is believed to have a trend. Using a large amount of data, this function outputs a vector that represents the Holt's method exponential smooth.
    Input: - data: a list of lists containing raw data to smooth using Holt's method
           - a: the weight given to the more recent data
           - b: The weight given to the older data
           - n: The number of months used for initialization
    Output: - ans: a vector that represents the Holt's method exponential smooth
    """
    #note this function assumes the data is a list of lists. Will encounter errors if the data is just a single list.
    Ls = [row[n-1] for row in data]#a running list of Lt, starting with the L0 being the final month of the initalization period
    Ts = [] #a running list of Tt values for each row to refer back to and run the method
    for row in data:
        T0 = []
        for i in range(n-1,0,-1): #should be counting backwards
            Tval = row[i] - row[i-1]
            T0.append(Tval) 
        top = sum(T0)
        bottom = len(T0)
        Ts.append(top/bottom) #this ideally outputs the T0 value 
    ans = [] #initialize the eventual output vector
    for row in data:
        #index based on the right
        for i in range(n+1,len(row)+1):
            Lt = a*row[i-1]+(1-a)*(Ls[i-n-1]+Ts[i-n-1])
            Ls.append(Lt)
            Tt = b*(Ls[i-n]-Ls[i-n-1])+(1-b)*Ts[i-n-1]
            Ts.append(Tt)
            ans.append(Ls[i-n-1]+Ts[i-n-1])
    return ans
            

holts([[4,6,8,10,14,18,20,22,24,28,31,34,40,47,50,49,56,53,55,63,68,65,72,69,79,82,80,85,94,89,96,100,100,105,108,110]],0.3,0.1,12)
```

```python
#lesson 10 Problem 2

import datetime as dt
#year = input("What is the Year: ")
#month = input ("What is the Month: ")
year = "2019"
month="10"

def calendar(year,month):
    """
    Input: year - string or integer - 4 digits
           month - string or integer of month
    Output: 
           6x7 nested list where the number in the list
           represents the date for the day, positioned so that
           all days of the same week line up under each other in order
           "" should be for positions that do not have a day
    """
    
    cal = [["" for i in range(7)] for j in range(6)]
   
    yr = int(year)
    mon = int(month)
    DAYS = ["Sunday","Monday","Tuesday","Wednesday","Thursday",
     "Friday","Saturday"]
    currDate = dt.datetime(yr,mon,1)
    startDay = DAYS.index(currDate.strftime("%A"))
    numDays = (dt.datetime(yr,mon+1,1)-currDate).days
   
    
    for day in range(1,numDays+1):
        position = startDay+day-1
        col = position//7
        row = position%7
        cal[col][row] = day
    return cal

cal = calendar(year,month)


def printCalendar(cal):
    DAYS = ["Sunday","Monday","Tuesday","Wednesday","Thursday",
     "Friday","Saturday"]
    res = ""
    res= "{:} {:} {:} {:} {:} {:} {:}\n".format(*DAYS)
    print(res)
    for i in range(len(cal)):
        res="{:>5} {:>6} {:>7} {:>8} {:>9} {:>7} {:>6}\n".format(*cal[i])
        print(res)
        
printCalendar(cal)
```

```python
# EX 3

def MA(ts,N):
    """
    MA(time_series,N) is a function that calculates the moving average of N
    days and returns a series with N-1 less elements than the original
    Inputs: time_series - a series of numbers that represent some measurement
            over the course of n discrete periods of time (days, minutes, ...)
            N - the number of elements to average
 
    For example say the series 1,5,3,2,6,5,4 was the series and N = 3,
    then (1+5+3)/3 = 3, (5+3+2)/3 = 3.333, (3+2+6)/3 = 3.667, (2+6+5)/3=4.333,
    (6+5+4)/3 = 5 => {3.000, 3.333, 3.667, 4.333, 5.000} is the output series.
    """
    
    m = len(ts)
    ma = []
    if N > m:
        print("Moving average not possible. Desired moving average length longer than vector.")
    for i in range(m-N+1):
        ma.append(sum(ts[i:i+N])/N)
    return ma

import random
data = [1,5,3,2,6,5,4]

print(MA(data,3))
```
