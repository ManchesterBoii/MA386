```python

ys = list(np.arange(0,1.1,0.2))
xs = ['G']+list(range(5,50,5)) 
xs += [50] + xs[::-1] 

red = pd.Series([13,18,19,22,21,25,28,29,32,38,41,42,49,54,60,58,68,72,71,87] )
blue = pd.Series([18,20,22,28,28,30,33,36,41,45,50,53,57,71,73,73,82,89,82,91])
#plt.xticks(xs,("5", "10", "15", "20", "25", "30", "35", "40", "45", "50", "45", "40", "35", "30", "25", "20", "15", "10", "5", "G"))
 
#plt.plot(red.index,red.values)
vals = list(range(0,105,5))
x=range(len(vals))

red = red/100
blue = blue/100
x_k=range(-1,len(xs)-1)

s = 80
fig, ax = plt.subplots(figsize=(12, 4))
fig.set_dpi(100)
ax.plot(list(range(len(blue))), blue, red, color= "y")
plt.scatter(list(range(len(blue))), blue, red, color= "y")  
fig.set_facecolor('grey')
ax.set_facecolor('g')
lst = list(range(len(blue)))

for x,y in zip(lst,blue):

    plt.annotate(f"{y*100:.0f}%", # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(11,12), # distance from text to points (x,y)
                 ha='center',
                 color = "y")
for x,y in zip(lst,red):

    plt.annotate(f"{y*100:.0f}%", # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(11,-12), # distance from text to points (x,y)
                 ha='center',
                 color ="y")
       
plt.xticks(x_k, xs ) 
plt.xlim(-5,22)
plt.ylim(0,1)

for i in range(0,len(red)):
    ax.axvline(x=i, color= 'w')

ax.annotate("Efficiency = Score/Possession", xy=(-1,0.8),color= "yellow",fontsize=12)
for i , alp in enumerate('TOUCHDOWN'):
    ax.annotate(alp, xy=(19.8, 0.9-(i/10)), color ='white', fontsize=12)

fig.canvas.draw() 
plt.show() 
```
