```python
def txtToDict(filename):
    """
    Input-  filename - string that specifies the file name 
    
    Output: This function produces a nested dictionary that is a data structure.
            The keys (which are the course numbers) in the outside dictionary map to three dictionaries.
            One of these dictionaries maps 'Course Name' to the individual course name
            One of these dictionaries maps "Credit" to the idividual course credit
            One of these dictionaries maps 'Grade' to the individual course grade
            The first header columns name
    """
    with open(filename, mode='r') as input_file:
        data = {}
        header = str(input_file.readline()).strip().split(",")
        course = {}
        grade = {}
        for line in input_file:
            lists = line.strip().split(",")
            data[lists[0].replace(" ", "")] = { header[1]: lists[1], header[2] : lists[2], header[3]:lists[3]  }
            #data[line[0]]=line[1:]
    return header[0],data
data = txtToDict('student.csv')
data
#print(data)
```

```python
def dictToTxt(dat, outfile, fmt = ".csv"):
    '''
    Takes a dictionary and writes it to a csv
    
    Inputs: dat - tuple containing the header title of the first data column and a dictionary
            outfile - name of txt file we are writing to
    Outputs: None, writes to specified text file
    fmt - specifies .txt for .csv file
    '''
    
    header0, d = dat
    keys = []
    vals = []
    form = 0
    for i,j in d.items(): #use this because d.keys() gives us a tuple, not good
        keys.append(i)
        vals.append(j)
    if isinstance(vals[0], dict):
        form = 1
        header = []
        for elem in vals[0].keys():
            header.append(elem)
        for i in range(len(vals)):
            str_repr = keys[i] + ','
            for elem in vals[i].values():
                str_repr += str(elem)
                str_repr += ','
            vals[i] = str_repr[:-1]
    if form == 1:
        keys = header
        keys = [header0] + keys
    vals = np.array(vals).T.tolist() #takes transpose (rotates data to write easier)
    writecsv(header = keys, data = vals, csvName = outfile, decimals = 3, fmt = fmt)

d = ("name",{'Radford': {'Class': 'firstie', 'Company': 'B1','Age': 22} , 'Andrew': {'Class': 'firstie', 'Company': 'F1','Age': 21}, 'Vedin': {'Class_Name': 'cow', 'Company': 'B1','Age': 20}, 'Nzia': {'Class_Name': 'yuk', 'Company': 'C1','Age': 22}})

dictToTxt(d, 'cadets', fmt=".csv")

data = txtToDict('student.csv')
dictToTxt(data, 'student2', fmt= ".txt")
```
